# Getting Started with Regress-LM

This tutorial provides a step-by-step walkthrough of how to use Regress-LM for a simple regression task. We will predict house prices from text descriptions.

```python
from regress_lm import core
from regress_lm import rlm
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Create Examples

First, we create a few example (x, y) pairs to fine-tune our model. The input `x` is a text description of a house, and the output `y` is its price.

```python
examples = [
    core.Example(x='A charming 2-bedroom bungalow with a spacious garden.', y=300000.0),
    core.Example(x='A modern 5-bedroom mansion with a swimming pool.', y=1200000.0),
    core.Example(x='A small studio apartment in the city center.', y=150000.0),
    core.Example(x='A 3-bedroom house in the suburbs with a garage.', y=450000.0),
    core.Example(x='A luxurious penthouse with stunning city views.', y=2000000.0)
]
```

## 2. Initialize and Fine-Tune the Model

Next, we initialize a Regress-LM model from scratch and fine-tune it on our examples.

```python
reg_lm = rlm.RegressLM.from_scratch(max_input_len=2048)
reg_lm.fine_tune(examples)
```

## 3. Make Predictions

Now that the model is fine-tuned, we can make predictions on new, unseen house descriptions.

```python
queries = [
    core.ExampleInput(x='A cozy 1-bedroom apartment with a balcony.'),
    core.ExampleInput(x='A large 4-bedroom family home with a big yard.')
]

samples1, samples2 = reg_lm.sample(queries, num_samples=128)
```

## 4. Visualize Results

Finally, we visualize the distribution of predictions for each query.

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(samples1, bins=20, alpha=0.7)
plt.title('Predictions for: "A cozy 1-bedroom apartment..."')
plt.xlabel('Predicted Price')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(samples2, bins=20, alpha=0.7)
plt.title('Predictions for: "A large 4-bedroom family home..."')
plt.xlabel('Predicted Price')

plt.tight_layout()
plt.show()
```
